```python
"""
Hang Man 

The Game : The computer randomly chooses a word from a given set of words (here a set of mathematics related words) 
           and displays each letter as a blank to the user. The user has to guess each letter of the word. The 
           maximum admissible number of incorrect guesses is 6, after which the user will be given a final chance to 
           guess the word as a whole.
Note : The user won't be penalised for wrongly guessing a letter more than once, instead will be shown a message 
       regarding the same.
"""
```

```python
words=["abelian group", "abridged multiplication", "abscissa", "absolute error", "absolutely convergent series", "absolute maximum", "accumulation point", "acute angle", "algebraic expression", "alternating series", "antiderivative", "antisymmetric matrix", "arc length", "arithmetic mean", "arithmetic progression", "augmented matrix", "auxiliary equation", "axiom", "barycentre", "bernoulli distribution", "bessels equation", "bijection", "bilinear", "binary operation", "binomial coefficients", "boundary conditions", "calculus", "canonical embedding", "cardinal number", "cartesian coordinate system", "cauchy riemann equations", "central limit theorem", "central tendency", "characteristic equation", "circumference", "classical mechanics", "coefficient matrix", "denominator", "dependent variable", "determinant", "differentiation", "diophantine equation", "disconnected graph", "eccentricity", "eigenvalue", "euclidean algorithm", "factorisation", "fermat numbers", "fibonacci sequence", "fudamental theorem of calculus", " guassian elimination", "general term", "greatest common divisor", "harmonic mean", "hemisphere", "hermitian matrix", "hypotenuse", "identity element", "imaginary number", "independent variable", "inequality", "inner product", "jump discontinuity", "kilogram", "laplacian", "lemma", "linear algebra", "lower bound", "magnitude", "matrix", "mean value theorem", "multiplicity", "natural logarithm", "nonlinear", "norm", "numerator", "odd number", "orthogonal", "projection", "pythagoras theorem", "quartile", "radius", "symmetry", "triangles", "trigonometric equations", "uncountable set", "venn diagram", "vector space", "wronskian", "zero"]

import random
word = random.choice(words)

#word = "algebraic topology"
l_word = len(word)

listed = list(word)
```

```python
space_index = []
for i in range(l_word):
	if word[i] == " ":
		space_index.append(i)
```

```python
x = []
for i in range(l_word):
	if i in space_index:
		x.append(" ")
	else:
		x.append("_ ")

def display(x):
	dis=""
	for d in range(l_word):
		dis += x[d]
	return(dis)
```

```python
fig = 0

figure(fig)
print(display(x))

incorrect_guesses = []
num_incorrect = 0

while num_incorrect < 6:
	print("\n\n________________________________________\n\n")
	guess = input("Enter your letter : ")
	
	if guess not in listed:
		
		if guess in incorrect_guesses:
			print("\nThis letter was already guessed and was found to be incorrect.")
		
		else:
			fig += 1
			num_incorrect += 1
			print("\nIncorrect guess.")
		figure(fig)
		incorrect_guesses.append(guess)
	
	else:
		
		for i in range(l_word):
			if guess == word[i]:
				x[i] = word[i]
		
		if display(x) == word:
			figure("happy_face")
			print(display(x))
			print("\nCongratulations! Correct guess.")
			figure("game_over")
			break
		else:
			print("\nCorrect guess.")
			figure(fig)
	print(display(x))
	print("\nNumber of incorrect guesses left : ", 6 - num_incorrect)
	
	if 6 - num_incorrect == 0:
		guess_word = input("You've exhausted all the chances. Guess your word : ")
		
		if guess_word == word:
			figure("happy_face")
			print("\nCongratulations! Correct guess.")
			figure("game_over")
			break
		else:
			figure("dead_face")
			print("\nIncorrect guess.\nThe correct word is {}.\nBetter luck next time.".format(word))
			break
			
```
